```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("test.csv")
x = np.array(data.iloc[:,0])
y = np.array(data.iloc[:,1])
```

```python
class linearregression:
    def __init__(self,x,y):
        self.x = x
        self.y = y
        self.m = 0
        self.b = 0
        self.n = len(x)
       
    def fit(self,epochs,ls):
        #Implementing Gradient Descent
        for i in range(epochs):
                y_pred   = self.m * x + self.b
        
              #Calculating derivatives w.r.t Parameters
                D_m = (-2/self.n)*sum(self.x * (self.y - y_pred))
                D_b = (-1/self.n)*sum(self.y-y_pred)
        
               #Updating Parameters
                self.m = self.m - ls*D_m
                self.b = self.b -ls *D_b
     
    def mse(self,inp_y,inp_ypred):
        return f"MSE ==> {(1/self.n)*sum(inp_y - inp_ypred)**2}"
        
    def predict(self,inp):
        y_predict = self.m * inp + self.b
        return y_predict   
        
```

```python
model = linearregression(x,y)
```

```python
#Training the model with .fit method
model.fit(1000,.0001)
```

```python
#Prediciting the values
y_pred = model.predict(x)
```

```python
score=model.mse(y,y_pred)
print(score)
```

```python
#Plotting the results
plt.figure(figsize = (10,6))
plt.scatter(x,y , color = 'green')
plt.plot(x , y_pred , color = 'k' , lw = 3)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
